```python
#require packages
from ragraph.io.csv import from_csv
import ragraph.plot

import plotly.graph_objs as go

#block function
def add_dsm_blocks(
    fig,
    blocks,
    n_nodes,
    xref,
    yref,
    layer="below",
):
    """
    blocks: list of dicts with keys
        row_start, row_end  (inclusive, 0-based)
        col_start, col_end  (inclusive, 0-based)
        color
        opacity (optional)
        line_color (optional)
        line_width (optional)
    """

    shapes = []

    for block in blocks:
        r0, r1 = block["row_start"], block["row_end"]
        c0, c1 = block["col_start"], block["col_end"]

        shape = go.layout.Shape(
            type="rect",
            # X spans columns
            x0=c0,
            x1=c1 + 1,
            # Y spans rows (DSM is inverted)
            y0=n_nodes - r1 - 1,
            y1=n_nodes - r0,
            fillcolor=block["color"],
            opacity=block.get("opacity", 0.3),
            line=dict(
                color=block.get("line_color", "black"),
                width=block.get("line_width", 1),
            ),
            xref=xref,
            yref=yref,
            layer=layer,
        )

        shapes.append(shape)

    fig.update_layout(shapes=fig.layout.shapes + tuple(shapes))

```

```python
# Converting csv files into a graph object.
course = from_csv(
    nodes_path="nodes_course.csv", #name of nodefile
    edges_path="edges_course.csv", #name of edge file
    csv_delimiter=",",
    iter_delimiter=","  # Separates list elements within a cell.
)

#common errors:
#-key error: likely the edges contain a name that does not appear in the nodes file. often a typo.

#build the dsm, 
dsm = ragraph.plot.dsm(
    leafs = course.nodes,
    edges = course.edges,
    style= ragraph.plot.Style(
        piemap=dict(
            display="labels",
            mode="relative",
            fields=course.edge_labels  #sorted by edge labels, so numerical order
        ),
        palettes=dict(
            fields={
                "EA flow": {"categorical": "#E57373"},  # Soft red, easy on the eyes
                "chemical flow": {"categorical": "#F49600"},  # Warm orange with good visibility
                "tray flow": {"categorical": "#2A642E"},  # Light green: gentle, contrasting with others
                "substrate flow": {"categorical": "#66BB6A"},  # Slightly darker green for better contrast
                "measurement flow": {"categorical": "#609BCC"},  # Soft blue: clean and calm
                "control flow": {"categorical": "#D12C8A"},  # Softer yellow (pale gold)
            }
        )

    )
)

blocks = [
    dict(
        row_start=0,
        row_end=6,
        col_start=0,
        col_end=6,
        color="rgba(0, 0, 255, 0.1)",
    ),
    dict(
        row_start=0,
        row_end=9,
        col_start=7,
        col_end=9,
        color="rgba(255, 0, 0, 0.7)",
    ),
    dict(
        row_start=7,
        row_end=9,
        col_start=0,
        col_end=6,
        color="rgba(255, 0, 0,0.7)",
    ),
        dict(
        row_start=3,
        row_end=4,
        col_start=3,
        col_end=4,
        color="rgba(0, 255, 0, 1)",
    ),
]

add_dsm_blocks(
    dsm,
    blocks,
    n_nodes=len(course.leafs),
    xref="x9",
    yref="y9",
)


dsm
```
